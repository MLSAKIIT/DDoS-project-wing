```python
def pre_process(df,col):
    #df['primary_use'] = df['primary_use'].astype("category").cat.codes
    print("Name of the column name with NaN: "+str(col))
    print(df[col].value_counts(dropna=False,normalize=True).head())
    df[col].replace(np.inf, -1, inplace=True)
    
    return df
```

```python
#memory usage reduction

def convert(df):
    start_mem_usage = df.memory_usage().sum()/1024**2
    #NAN list = [] keeping tracks of columns with missing values
    print("Memory usage of properties dataframe is:", start_mem_usage,"MB")
    for col in df.columns:
        if df[col].dtype!=object:  #exclude strings
            if df[col].dtype==float:
                df[col] = df[col].astype(np.int)
                
            print("dtype after: ", df[col].dtype)
            
    print("Memory usage after completion:- ")
    mem_usg=df.memory_usage().sum()/1024**2

    print("Memory usage is: ", mem_usg,"MB")
    print("This is ",100*mem_usg/start_mem_usage,"% of the initial size")
    return df

convert(df)
```

CHI-SQUARED TEST

```python
#H0: there exists no relation between two features
#H1: there exists some relation between two features features under a condition
```

```python
from scipy.stats import chi2_contingency

#between ddos flag count and protocol
stat,p,dof,expected=chi2_contingency(pd.crosstab(df['Label']==1,df['Protocol']))
alpha=0.05
print("p value is "+str(p))
print("ddos","Protocol")
if p>alpha:
    print("Independent(H0 holds true)")
else:
    print("Dependent(H1 holds true)")
    
    
#between benign flag count and protocol    
stat,p,dof,expected=chi2_contingency(pd.crosstab(df['Label']==0,df['Protocol']))
alpha=0.05
print("p value is "+str(p))
print("Benign","Protocol")
if p>alpha:
    print("Independent(H0 holds true)")
else:
    print("Dependent(H1 holds true)")
```

Principal Component analysis

```python

from sklearn.preprocessing import StandardScaler
x=df.drop(['Label'], axis=1)
y=df['Label']
if np.all(np.isfinite(x)) == False:
    x_std = StandardScaler().fit_transform(df)
    mean_vec=np.mean(x, axis=0)
    cov_mat = (x-mean_vec).T.dot((x-mean_vec))/(x.shape[0]-1)
    
```

```python
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
print("Eigenvectors \n%s" %eig_vecs)
print("Eigenvalues \n%s" %eig_vals)
```

```python
eig_pairs = [(np.abs(eig_vals[i]),eig_vecs[:,i]) for i in range(len(eig_vals))]

#sorting the values from high to low
eig_pairs.sort(key=lambda x: x[0], reverse=True)

#visual confirmation of the sorting
print("Descending order:")
for i,j in enumerate(eig_pairs):
    print(i,j[0])
    
```

Explained variance

```python
tot= sum(eig_vals)
var_exp= [(i/tot)*100 for i in sorted(eig_vals, reverse=True)]
```

```python
with plt.style.context('dark_background'):
    plt.figure(figsize=(30,30))
    
    plt.bar(range(78), var_exp, alpha=0.5, align='center', label='individual explained variance')
    
    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal componenets')
    plt.legend(loc='best')
    plt.tight_layout()
```
